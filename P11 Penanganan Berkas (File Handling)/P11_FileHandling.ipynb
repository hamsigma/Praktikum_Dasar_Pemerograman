# **Modul Pertemuan 12: Penanganan Berkas (File Handling)**

## **Pembahasan**

### **1. Apa itu Penanganan Berkas?**

Penanganan berkas adalah proses melakukan operasi pada berkas di sistem penyimpanan, seperti membuka, membaca, menulis, dan menghapus berkas. Dalam Python, kita dapat menggunakan fungsi built-in seperti `open()`, `read()`, `write()`, dan `close()` untuk menangani berkas.

### **2. Membuka Berkas**

Untuk membuka berkas, kita menggunakan fungsi `open()`, yang memerlukan nama berkas dan mode akses (misalnya, "r" untuk membaca, "w" untuk menulis, dan "a" untuk menambahkan).

Contoh membuka berkas:
```python
file_handling = open("contoh.txt", "r")
```

### **3. Membaca Berkas**

Setelah membuka berkas, kita dapat membaca isinya menggunakan beberapa metode:

- **Membaca seluruh isi berkas:**
    ```python
    isi_teks = file_handling.read()
    print(isi_teks)
    ```

- **Membaca baris per baris:**
    ```python
    for baris in file_handling:
        print(baris)
    ```

- **Membaca dalam bentuk list:**
    ```python
    list_baris = file_handling.readlines()
    ```

### **4. Menulis Berkas**

Untuk menulis ke dalam berkas, kita membuka berkas dengan mode "w" atau "a".

Contoh menulis ke dalam berkas:
```python
file_handling = open("contoh.txt", "w")
file_handling.write("Contoh 1\n")
```

Jika kita ingin menambahkan data tanpa menghapus data yang ada, kita menggunakan mode "a":
```python
file_handling = open("contoh.txt", "a")
file_handling.write("Data baru\n")
```

### **5. Menutup Berkas**

Setelah selesai melakukan operasi pada berkas, kita harus menutup berkas tersebut dengan metode `close()`:
```python
file_handling.close()
```

````python
import os

# Class untuk menyimpan data mahasiswa
class DataMahasiswa:
    def __init__(self, nama, nim):
        self.nama = nama
        self.nim = nim

    def simpan_data(self):
        with open("data_mahasiswa.txt", "a") as file:
            file.write(f"{self.nama}, {self.nim}\n")
            print("Data berhasil disimpan ke dalam data_mahasiswa.txt.")

# Class untuk membaca data mahasiswa
class BacaData:
    def baca_data(self):
            with open("data_mahasiswa.txt", "r") as file:
                data = file.readlines()
                if not data:
                    print("File tidak ada yang di tampilkan.")
                    return
                for line in data:
                    nama, nim = line.strip().split(", ")
                    print(f"Nama Lengkap: {nama}, NIM: {nim}")

# Class untuk mengelola daftar kehadiran
class Kehadiran:
    def __init__(self, nama, status):
        self.nama = nama
        self.status = status

    def simpan_kehadiran(self):
        if self.cek_nama():
            print("Nama sudah ada di daftar.")
            return
        with open("kehadiran.txt", "a") as file:
            file.write(f"{self.nama} - {self.status}\n")
            print("Kehadiran berhasil disimpan.")

    def tampilkan_kehadiran(self):
            with open("kehadiran.txt", "r") as file:
                data = file.readlines()
                if not data:
                    print("Tidak ada data kehadiran.")
                    return
                for line in data:
                    print(line.strip())

    def cek_nama(self):
            with open("kehadiran.txt", "r") as file:
                data = file.readlines()
                for line in data:
                    if self.nama in line:
                        return True
       
# Class untuk menghapus file
class FileManager:
    def hapus_file(self, nama_file):
        konfirmasi = input(f"Apakah Anda yakin ingin menghapus file {nama_file}? (y/n): ").lower()
        if konfirmasi == "y":
                os.remove(nama_file)
                print(f"File {nama_file} berhasil dihapus.")
                
def menu():
    while True:
        print("\nPilihan:")
        print("1. Simpan Data Mahasiswa")
        print("2. Baca Data Mahasiswa")
        print("3. Simpan Data Kehadiran")
        print("4. Tampilkan Data Kehadiran")
        print("5. Hapus File")
        print("6. Keluar")
        
        pilihan = input("Pilih menu (1-6): ")
        
        if pilihan == "1":
            nama = input("Masukkan Nama Lengkap: ")
            nim = input("Masukkan NIM: ")
            mahasiswa = DataMahasiswa(nama, nim)
            mahasiswa.simpan_data()
        
        elif pilihan == "2":
            bacadata = BacaData()  # Membuat objek dari class BacaData
            bacadata.baca_data()
        
        elif pilihan == "3":
            nama = input("Masukkan Nama Lengkap: ")
            status = input("Masukkan Status Kehadiran (Hadir/Tidak Hadir): ")
            kehadiran = Kehadiran(nama, status)
            kehadiran.simpan_kehadiran()
        
        elif pilihan == "4":
            kehadiran = Kehadiran("", "")  # Membuat objek kosong untuk menampilkan kehadiran
            kehadiran.tampilkan_kehadiran()
        
        elif pilihan == "5":
            file = input("Masukkan nama file yang ingin dihapus (misalnya kehadiran.txt atau data_mahasiswa.txt): ")
            file_manager = FileManager()  # Membuat objek FileManager
            file_manager.hapus_file(file)
        
        elif pilihan == "6":
            print("Keluar dari program.")
            break
        
        else:
            print("Pilihan tidak valid. Coba lagi.")

# Memulai program
if __name__ == "__main__":
    menu()
````
